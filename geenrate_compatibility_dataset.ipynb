```python
import pandas as pd 
from scipy.stats import pearsonr, spearmanr, kendalltau
import numpy as np
batch_size=4096
```

```python
experiment_id='c'
res_path = f"/nfs/scistore19/locatgrp/dyao/multiview-crl-eval/results/numerical/five_latents_{experiment_id}_all_data.csv"
df = pd.read_csv(res_path)
```

```python
save_path=f"/nfs/scistore19/locatgrp/dyao/causal-self-compatibility/data/T-MEX/model_{experiment_id}"
```

```python
for i in range(50):
    df_i = df.iloc[i*batch_size:(i+1)*batch_size]
    save_file_path = f"{save_path}/dataset{i}.csv"
    df_i = df_i.rename(columns={'z0': 'V1', 'z1': 'V2', 'z2': 'V3', 'z0_est0': 'V4'})
    df_i.to_csv(save_file_path, index=False, columns=['V1', 'V2', 'V3', 'V4'])
```

```python
spearman_correfs = []
```

```python
for batch_id in range(50):
    zs = df.iloc[batch_id*batch_size:(batch_id+1)*batch_size][['z0', 'z1', 'z2']].values
    z_hat = df.iloc[batch_id*batch_size:(batch_id+1)*batch_size][['z0_est0']].values
    spearman_correfs.append(np.abs(spearmanr(z_hat.squeeze(), zs).statistic[0, 1:]))
```

```python
spearman_correfs = np.stack(spearman_correfs)
```
